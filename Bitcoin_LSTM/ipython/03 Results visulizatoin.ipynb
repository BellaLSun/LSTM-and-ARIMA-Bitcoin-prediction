```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import pandas as pd
```

#### 7 days, single point prediction

```python
RMSE_on_90_train_days = [
414.952379,
693.787074,
113.54812,
212.922658,
149.373661]

RMSE_on_120_train_days = [
285.103577,
501.798944,
151.803832,
221.173472,
172.285768    
]

RMSE_on_270_train_days = [
280.69896,
454.138756,
177.017771,
211.050172,
138.16067    
]
```

```python
time = ['2018-6-18','2018-7-18','2018-8-18','2018-8-24','2018-9-14']
```

```python
single_point_prediction = {
    "time":time,
    "RMSE_on_90_train_days":RMSE_on_90_train_days,
    "RMSE_on_120_train_days":RMSE_on_120_train_days,
    "RMSE_on_270_train_days":RMSE_on_270_train_days
}
single_point_prediction = pd.DataFrame(single_point_prediction)
# single_point_prediction = single_point_prediction.set_index('time')
# single_point_prediction.index = pd.to_datetime(single_point_prediction.index)
single_point_prediction
```

```python
plt.figure(figsize=(12,8))
# .plot(style='-.bo')

fig,ax = plt.subplots()
# single_point_prediction['RMSE_on_90_train_days'].plot(style='-.bo')
# single_point_prediction['RMSE_on_120_train_days'].plot(style='-.bo')
# single_point_prediction['RMSE_on_270_train_days'].plot(style='-.bo')

plt.plot(single_point_prediction['time'],single_point_prediction['RMSE_on_90_train_days'])
plt.plot(single_point_prediction['time'],single_point_prediction['RMSE_on_120_train_days'])
plt.plot(single_point_prediction['time'],single_point_prediction['RMSE_on_270_train_days'])
#设置数字标签**
for a,b in zip(single_point_prediction['RMSE_on_90_train_days'].index,single_point_prediction['RMSE_on_90_train_days']):
    ax.text(a, b, '%.4f' % b, ha='center', va= 'bottom',fontsize=9)

for a,b in zip(single_point_prediction['RMSE_on_120_train_days'].index,single_point_prediction['RMSE_on_120_train_days']):
    ax.text(a, b, '%.4f' % b, ha='center', va= 'bottom',fontsize=9)

for a,b in zip(single_point_prediction['RMSE_on_270_train_days'].index,single_point_prediction['RMSE_on_270_train_days']):
    ax.text(a, b, '%.4f' % b, ha='center', va= 'bottom',fontsize=9)

ax.set_title('7 days single point prediction',fontsize=16)
plt.legend(bbox_to_anchor=(1.0, 1), loc=1, borderaxespad=0.)
plt.show()
```

### Conclusion

- **Single point prediction**
- more days get trained, the RMSE will be smaller, the model will be more accurate.
- when the data is rather stationary, the RMSE does not have too much difference over different training days.


- **Time sequence prediction of 1 LSTM layer**
- The smaller the windown size is, the more accurate the model will be.


- **Different input dimensions**
- No big difference between 6 dimensions and 2 dimensions after combining features, which proves that deep learning model is robust to the data with multicollinearity property.
- 6-dimensions model is more accurate than 2-dimensions. 


- **Stateful LSTM**
- one layer LSTM performs better than three layers, the reason might be that the price of Bitcoin is volatile therefore the more neurons and layers will cause overfitting and less accurate on test dataset prediction.
